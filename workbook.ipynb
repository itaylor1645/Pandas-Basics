```python
#Create a feature that is First Name + Last Name
df["full_name"]=df["first_name"] + " "+ df["last_name"]
df.head()
```

```python
#create a feature where age < 30 is young, 30<=50 is mid, and older is senior
df["age_group"]=np.where(
    df["age"] < 30, "Young",
    np.where(df["age"] <=50, "Mid", "Senior")
)
df.head()
```

```python
#count people in each age group
df["age_group"].value_counts()
```

```python
#create a feature that contains just the email_domain-
df["email_domaine"]=df["email"].str.split("@").str[1]
df.head()
```

```python
#check which email domain is most common
df["email_domaine"].value_counts().idxmax()
```

```python
#Find all people whose email ends with .com
df[df["email"].str.endswith(".com")]
```
